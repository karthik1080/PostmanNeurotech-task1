```python
from scipy.signal import welch
import matplotlib.pyplot as plt
import numpy as np
eeg_file_path = "C:/Users/karth/Desktop/eeg-data.txt"
with open(eeg_file_path, 'r') as file:
    eeg_data = np.array([float(line.strip()) for line in file])


delta = (1, 4)
theta = (4, 8)
alpha = (8, 13)
beta = (13, 30)

# Compute power spectral density
fs = 100  
f, psd = welch(eeg_data, fs=fs, nperseg=1024)

# Calculate absolute bandpowers
deltap = sum(psd[(f >= delta[0]) & (f <= delta[1])])
thetap = sum(psd[(f >= theta[0]) & (f <= theta[1])])
alphap = sum(psd[(f >= alpha[0]) & (f <= alpha[1])])
betap = sum(psd[(f >= beta[0]) & (f <= beta[1])])

# Calculate total power
totalp = (deltap+thetap+alphap+betap)

# Calculate relative bandpowers
reldelta = deltap / totalp
reltheta = thetap / totalp
relalpha = alphap / totalp
relbeta = betap / totalp

print(f"Delta relative bandpower: {reldelta:.4f}")
print(f"Theta relative bandpower: {reltheta:.4f}")
print(f"Alpha relative bandpower: {relalpha:.4f}")
print(f"Beta relative bandpower: {relbeta:.4f}")
max_value = max(reldelta,reltheta,relalpha,relbeta)
if max_value ==reldelta:
    print("the band with the highest relative power is reldelta")
elif max_value == reltheta:
    print("the band with the highest relative power is reltheta")
elif max_value == relalpha:
    print("the band with the highest relative power is relalpha")
else:
    print("the band with the highest relative power is relbeta")




# Plot frequency vs. PSD
plt.figure(figsize=(10, 6))
plt.plot(f, psd)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power/Frequency (dB/Hz)")
plt.title("EEG Power Spectral Density")
plt.show()

```
